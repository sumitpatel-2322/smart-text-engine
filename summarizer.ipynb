```python
import nltk
nltk.download("punkt")
from nltk.tokenize import sent_tokenize
```

```python
def split_into_sentences(text):
    sentences=sent_tokenize(text)
    return sentences
```

```python
if __name__=="__main__":
    text="""The movie was really slow at first.
    In the second half though, it started with a bang.
    The plot twist was engaging and interesting to watch.Towards the end it was totally worth it Well made very good."""
    sentences=split_into_sentences(text)
    for sentence in sentences:
        print(sentence)
```
